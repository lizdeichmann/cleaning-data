## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data
cols = df.columns
colors = ['#aba7a7', '#990800'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colors))
```

```python
# Try out a missing data percentage list! 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Plot a missing data histogram
# first create missing indicator for features with missing data
for col in df.columns:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        df['{}_ismissing'.format(col)] = missing


# then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in df.columns if 'ismissing' in col]
df['num_missing'] = df[ismissing_cols].sum(axis=1)

df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

There are many variables with close to or all data missing: Intersection Street 1, Intersection Street 2, Landmark, Facility Type, School or Citywide Complaint,  Vehicle Type, etc. As these variables can't provide any information, I will delete these variables. 

There are also several variables with considerably fewer missing values: Incident Address, Lattitude, Longitude, Location. For Incident Address and Street Name, which has 9% missing data,  I will replace missing data with the mode.

<br>

```python
# Handle any missing data in your dataframe.
#Drop variables
cols_to_drop = ['Intersection Street 1', 'Intersection Street 2', 'School or Citywide Complaint', 'Vehicle Type',
               'Taxi Company Borough', 'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction',
               'Road Ramp', 'Bridge Highway Segment', 'Garage Lot Name', 'Ferry Direction', 'Ferry Terminal Name',
               'Facility Type', 'Landmark']
df = df.drop(cols_to_drop, axis=1)
```

```python
# For missing values in "Incident Address" replace with mode
address_mode = df['Incident Address'].mode()
print(address_mode)
df['Incident Address'] = df['Incident Address'].fillna(df['Incident Address'].mode()[0])
```

```python
# For missing values in "Street Address" replace with mode
street_mode = df['Street Name'].mode()
print(street_mode)
df['Street Name'] = df['Street Name'].fillna(df['Street Name'].mode()[0])
```
